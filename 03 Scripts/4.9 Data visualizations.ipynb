# 03. Creating visualizations

```python
# Creating bar chart

ords_prods['orders_day_of_week'].value_counts().plot.bar()
```

```python
# Sorting the chart by day of week

bar = ords_prods['orders_day_of_week'].value_counts().sort_index().plot.bar()
```

```python
# Changing colors

bar = ords_prods['orders_day_of_week'].value_counts().plot.bar(color =['red', 'orange', 'yellow', 'green', 'blue', 'indigo', 'purple'])
```

```python
# Saving the visualization as png

bar.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_orders_dow.png'))
```

```python
# Creating a histogram

ords_prods['prices'].plot.hist(bins = 25)
```

```python
# Creating a scatterplot

sns.scatterplot(x = 'prices', y = 'prices',data = ords_prods)
```

```python
# Creating a more detailed histogram

hist_2 = ords_prods['prices'].plot.hist(bins = 70)
```

```python
# Saving the histogram as png

hist_2.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'hist_prices_freq.png'))
```

# 04. Sampling data 

```python
# Creating a 70/30 sample

np.random.seed(4)
dev = np.random.rand(len(ords_prods)) <= 0.7
```

```python
dev
```

```python
# Creating a list for the 70% of the dataframe

big = ords_prods[dev]
```

```python
# Creating a list for the 30% of the dataframe

small = ords_prods[~dev]
```

```python
# Checking if the split was correct

len(ords_prods)
```

```python
len(big) + len(small)
```

```python
# Narrowing down the number of columns to the essential ones

df_2 = small[['orders_day_of_week','prices']]
```

```python
df_2.head()
```

```python
line = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
```

```python
# Saving the line chart as png

line.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line_prices_per_day.png'))
```
