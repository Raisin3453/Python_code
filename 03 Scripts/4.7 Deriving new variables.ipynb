# 03. Product sorting

```python
# Defining the dataframe for only the first 1M rows
df = ords_prods_merge[:1000000]
```

```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
df['prices'].max()
```

```python
df.describe ()
```

# 04. Using loc() in the function

```python
# Creating conditions
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

# 05. For-loop if-statements

```python
# Renaming the order_dow
ords_prods_merge.rename(columns = {'order_dow':'orders_day_of_week'}, inplace = True)
```

# 06. Redoing the busiest days column

```python
# identifying the two days of the week for busyness

result = []

for value in ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 1:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  elif value == 3:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
ords_prods_merge['busiest_days'] = result
```

# 07. The busiest time of day column creation

```python
# Checking the frequency of order hours in order_hour_of_day

ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Creating a condition for the Most Orders for the busiest_period_of_day column

ords_prods_merge.loc[(ords_prods_merge['order_hour_of_day'] > 8) & (ords_prods_merge['order_hour_of_day'] < 17), 'busiest_period_of_day'] = 'Most Orders'
```

```python
# Creating a condition for the Fewest Orders for the busiest_period_of_day column

ords_prods_merge.loc[(ords_prods_merge['order_hour_of_day'] >= 0) & (ords_prods_merge['order_hour_of_day'] < 7), 'busiest_period_of_day'] = 'Fewest Orders'
```

```python
ords_prods_merge.loc[(ords_prods_merge['order_hour_of_day'] == 23), 'busiest_period_of_day'] = 'Fewest Orders'
```

```python
# Creating a condition for the Average Orders for the busiest_period_of_day column

ords_prods_merge.loc[(ords_prods_merge['order_hour_of_day'] > 6) & (ords_prods_merge['order_hour_of_day'] < 9), 'busiest_period_of_day'] = 'Average Orders'
```

```python
ords_prods_merge.loc[(ords_prods_merge['order_hour_of_day'] >= 17) & (ords_prods_merge['order_hour_of_day'] <= 22), 'busiest_period_of_day'] = 'Average Orders'
```
